#**Revenue Distribution by Booking Platform **

```python
df_new_datas.columns
```

```python
revenue_by_platform = df_new_datas.groupby("booking_platform")["revenue_realized"].sum().round(2).sort_values(ascending=False)
revenue_by_platform
```

```python
from matplotlib import pyplot as plt

revenue_by_platform = df_new_datas.groupby("booking_platform")["revenue_realized"].sum().round(2).sort_values(ascending=False)
total = revenue_by_platform.sum()

def autopct_func(pct):
    value = pct * total / 100
    return f'{pct:.1f}%\n{value:,.0f}'

plt.figure(figsize=(6,6))
plt.pie(
    revenue_by_platform,
    labels=revenue_by_platform.index,
    autopct=autopct_func,
    startangle=140,
    radius=0.7,
    colors=plt.cm.Set3.colors,
    textprops={'fontsize': 8}  # Reduce label and value font size
)
plt.title("Revenue Distribution by Booking Platform", fontsize=12)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
